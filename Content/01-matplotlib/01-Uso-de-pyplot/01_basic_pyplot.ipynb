```python
import matplotlib.pyplot as plt
import numpy as np
```

# Line charts

## *Change parameter*

>### Format strings 
>- ‘.’ ---> point marker
>- ‘,’ ---> pixel marker
>- ‘o’ --> circle marker
>- ‘v’ --> triangle_down marker
>- ‘^’ --> triangle_up marker
>- ‘<’ --> triangle_left marker
>- ‘>’ --> triangle_right marker

>### Line styles
>- ‘-’ --> solid line style
>- ‘–’ --> dashed line style
>- ‘-.’ --> dash-dot line style
>- ‘:’ ---> dotted line style

## *Charts*

```python
# make data
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y) # crea la gráfica
plt.show() # muestra la gráfica
```

```python
# Como tercer parámetro se pueden agregar caracteres que cambien el estilo de la gráfica
plt.plot(x,y,'r<:')
plt.show()
```

# Other charts

```python
plt.hist(x)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
# Relación entre dos distintas variables
plt.scatter(x,y)
plt.show()
```
