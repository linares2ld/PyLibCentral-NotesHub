```python
import matplotlib.pyplot as plt
import numpy as np
```

>## Parámetros -> plt.subplot()
>_plt.subplot(__filas__,__columnas__,__posición del arreglo__)_
>
>_

*__.subplot()__* permite tener más de una gráfica en un mismo espacio, dandole su lugar a cada gráfico.

```python
# make data
x = np.linspace(0,5,11)
y = x**2
```

```python
# Colocando la primer gráfica
plt.subplot(1,2,1)
plt.plot(x,y,'r--')

# Colocando la segunda gráfica
plt.subplot(1,2,2)
plt.hist(y)

# Mostrando las gráficas
plt.show()
```

```python
plt.subplot(1,2,1)
# Como ambos .plot() se encuentran debajo del mismo .subplot() compartiran el mismo espacio en el lienzo
plt.plot(x,y,'r--')
plt.plot(y,x,'m:')

plt.subplot(1,2,2)
plt.pie(y)

plt.show()
```

```python
# Se puede separar el lienzo por columnas o filas, todo dependerá de como se elijan los primeros 2
# parámetros de .subplot()
plt.subplot(2,1,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'m:')

plt.subplot(2,1,2)
plt.pie(y)
plt.show()
```
