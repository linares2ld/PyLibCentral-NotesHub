```python
import matplotlib.pyplot as plt
import numpy as np
```

># Programación Orientada a Objetos (OOP)
> Crear las gráficas de esta forma permite __mayor personalización__ que Pyplot

```python
# Data
x = np.linspace(0,5,11)
y = x**2
```

```python
# Creamos el lienzo donde trabajaremos
fig = plt.figure()

# Agregamos una gráfica (axes)
# Los parámetros de .add_axes() sirven para posicionar la gráfica dentro del lienzo. 
# En este caso no importan mucho pero son obligatorios.
axes = fig.add_axes([0.1,0.1,0.5,0.9])

# Le agregamos los datos a la gráfica
axes.plot(x,y,'b')
plt.show()
```

>## Parámetros -> .add_axes()
>*.add_axes([__posición_horizontal__,__posición_vertical__,__largo__,__alto__])*
>
>.

```python
# 2 objetos distintos que reposan en un mismo lienzo

# Creación del lienzo
fig = plt.figure()

# Agregamos las dos gráficas
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.55,0.4,0.3])

# Agregamos valores a ambas gráficas
axes.plot(x,y,'b')
axes2.plot(x,y,'w')

# Modificamos el estilo de la gráfica 2
axes2.set_facecolor('grey')

plt.show()
```
