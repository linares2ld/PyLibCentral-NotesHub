```python
import matplotlib.pyplot as plt
import numpy as np
```

Permite trabajar multiples gráficas en un solo lienzo con OOP

```python
# data
x = np.linspace(0,5,11)
y = np.sin(x)
```

>## Gráficas con subplot
>
> Código para una sola gráfica:
>- _fig,axes = plt.subplots()_
>
>Código para más de una gráfica:
>- _fig,axes = plt.subplots(nrows=__int__,ncols=__int__)_
>
>Los nombres de las variables no tienen que ser esos, pero son los nombres estandar


### Una gráfica

```python
fig,axes = plt.subplots()
axes.plot(x,y,'r')
plt.show()
```

### Más de una gráfica


- Ejemplo: 2 gráficas

```python
# Forma 1:
fig,axes = plt.subplots(nrows=1,ncols=2)

# Para acceder a un gráfico especifico es necesario ubicarlo como un objeto dentro de una lista
# Gráfica de la derecha
axes[0].plot(x,y,'r')
# Gráfica de la izquierda
axes[1].plot(y,x,'y')

plt.show()
```

```python
# Forma 2:

fig,(ax1,ax2) = plt.subplots(nrows=1,ncols=2)
# Los datos de la tupla tiene que coincidir con los parámetros de .subplots()

# En este caso, la segunda variable es una tupla con el nombre de cada gráfico,
# ahora solo se llama al gráfico por su nombre correspondiente
ax1.plot(x,y,'r')
ax2.plot(y,x,'y')

plt.show()
```

- Ejemplo: 8 gráficas

```python
# Forma 1:
fig,axes = plt.subplots(2,4)

axes[0][0].plot(x,np.cos(x))
axes[0][1].plot(x,np.sin(x),'y')
axes[0][2].plot(x,np.tan(x),'r')
axes[0][3].plot(x,np.cos(x)**2)

fig.tight_layout() # Acomoda las gráficas en automático para que se vean mejor
```

```python
# Forma 2:
fig,((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2,4)

ax1.plot(x,np.cos(x))
ax2.plot(x,np.sin(x),'y')
ax3.plot(x,np.tan(x),'r')
ax4.plot(x,np.cos(x)**2)

fig.tight_layout()
```
