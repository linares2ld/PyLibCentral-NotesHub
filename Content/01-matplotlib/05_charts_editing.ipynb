```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
x = np.linspace(0,5,11)
y = np.sin(x)
```

# Legends, labels, titles and sizes

## Usando Pyplot

```python
plt.plot(x,y,label='sin(x)')

# Editing
plt.title('Este es un título')
plt.xlabel('x')
plt.ylabel('y')
# 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
plt.legend(loc="upper right",bbox_to_anchor=(0.5, 0.5, 0.5, 0.5))

plt.show()
```

## Usando OOP

```python
# Con el parámetro figsize= se puede modificar el tamaño de los axes
fig,axes = plt.subplots(1,2,figsize=(10,4))

# Gráfica 1
axes[0].plot(x,y,label='$sin(x)$')

axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()

# Gráfica 2
axes[1].plot(y,x)

axes[1].set_title('Relacion Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')

plt.show()
```

# Colors and styles

__NOTA__: Cuando se aplica un estilo, este se aplicara a todas las gráficas del documento.

```python
# Default styles
print(plt.style.available)
```

## _Matlab structure_

```python
# Estilo del lienzo
plt.style.use('bmh')
fig,ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,'o')
ax.plot(x,x+2,'r--')
ax.plot(x,x+3,'g-')
ax.plot(x,x+4,'y.-')

plt.show()
```

## _Pyplot structure_

```python
plt.style.use('bmh')
fig,ax = plt.subplots(figsize=(8,8))

# alpha --> transparencia de la línea
# linewidth --> grosor de línea
# linestyle --> estilo de la línea
ax.plot(x,x+1,color='green',alpha=0.5,linewidth=5.,linestyle='--',marker='o',markersize=10,markerfacecolor='#FFF989')
ax.plot(x,x+2,color='#CED989',linewidth=10,linestyle='dashed',marker='x')
ax.plot(x,x+3,color='#66FF66',linestyle=':',marker='v')
ax.plot(x,x+4,color='#78CD00',marker='P')

plt.show()
```
