```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data:

# Categorical:
country = ['INDIA','JAPAN','MEXICO','COLOMBIA','GERMANY']

# Numeric:
population = [1000,800,900,1000,300]
data = np.random.randint(1,50,100)
```

# Column and bar
- Sirven para gráficar variables categóricas

```python
# Column
# Creación de la gráfica
plt.bar(country,population,width=0.5,color=['#789101','#123456'])
# Modificando las etiquetas categóricas
plt.xticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'),rotation=45)
plt.show()
```

```python
# Bar
plt.barh(country,population)
plt.show()
```

# Histograma

- Muestra la distribución de los datos

```python
# Parámetros:
# - width=
plt.hist(data,bins=30,histtype='step')
plt.show()
```

# Boxplot
They show:
- The median
- Upper and lower quartiles
- Minimum and maximum values
- Any outliers

```python
# Con este acción, podemos agregar un valor considerado outliner, para ver como se coomporta la gráfica
# Puede o no mostrar ese outliner con el valor del parámetro showfliers=
data = np.append(data,200)

plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=False)
plt.show()
data
```

# Scatter
- Muestra la relación entre dos variables

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N))**2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y,s=area,c=colors,marker='o',alpha=0.5)
plt.show()
```
